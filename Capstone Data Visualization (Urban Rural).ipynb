```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_all = pd.read_csv('Kaagle_Upload.csv')
df_all.head()
```

```python
df = df_all[['journey_purpose_of_driver', 'accident_severity', 'road_type', 'junction_detail', 'junction_control', 'light_conditions', 'weather_conditions', 'road_surface_conditions', 'special_conditions_at_site', 'carriageway_hazards', 'urban_or_rural_area', 'day_of_week', 'age_of_driver', 'age_of_vehicle']]
print(df.shape)
df.head()
```

```python
# remove all -1 values as they indicate data is missing
df = df[df['age_of_vehicle'] != -1]
df = df[df['age_of_driver'] != -1]
df = df[df['day_of_week'] != -1]
df = df[df['urban_or_rural_area'] != 3]
df = df[df['carriageway_hazards'] != -1]
df = df[df['special_conditions_at_site'] != -1]
df = df[df['road_surface_conditions'] != -1]
df = df[df['weather_conditions'] != -1]
df = df[df['light_conditions'] != -1]
df = df[df['junction_control'] != -1]
df = df[df['junction_detail'] != -1]
df = df[df['road_type'] != -1]
df = df[df['journey_purpose_of_driver'] != -1]
df = df[df['accident_severity'] != -1]
df.shape
```

### Urban or Rural area

```python
df_urac = df[['urban_or_rural_area', 'accident_severity']]
df_urac
```

```python
df_urac['urban_or_rural_area'].unique()
```

```python
df_urac['urban_or_rural_area'].replace(1, 'urban', inplace=True)
df_urac['urban_or_rural_area'].replace(2, 'rural', inplace=True)
df_urac
```

```python
df_urac3 = df_urac[df_urac['accident_severity'] == 3]
df_urac2 = df_urac[df_urac['accident_severity'] == 2]
df_urac1 = df_urac[df_urac['accident_severity'] == 1]

print(df_urac3['urban_or_rural_area'].unique())
print(df_urac2['urban_or_rural_area'].unique())
print(df_urac1['urban_or_rural_area'].unique())
```

```python
y3 = df_urac3['urban_or_rural_area'].value_counts()
df_uractot = pd.DataFrame({'accident severity 3': y3})
df_uractot
```

```python
y2 = df_urac2['urban_or_rural_area'].value_counts()
df_uractot.insert(1, 'accident severity 2', y2, True)
df_uractot
```

```python
y1 = df_urac1['urban_or_rural_area'].value_counts()
df_uractot.insert(2, 'accident severity 1', y1, True)
df_uractot
```

```python
df_uractot.plot(kind='bar', figsize=(10, 10))

plt.xlabel('Urban or rural area')
plt.ylabel('Number of Cases')
plt.title('Weather accidents happen in rural or urban areas')
plt.show()
```
